<span style="font-family: impact; font-size: 30px; color:#005b5b;"><b></b><u>HANDLING MISSING VALUES</span>

<li style="color:#c0c0c0; font-family:lucida"><i>The Columns [Brand, Category, Rating, Return_Rate] have Null values which are going to be handled as follows:</i> </li>

<span style="color:#4169e1; font-family: copperplate; font-size: 26px; font-weight: bold">a. Handling Brand Column:</span>

```python
recalldf['Brand'] = recalldf['Brand'].fillna("TMA")
recalldf.head(6)
```

<li style="color:#c0c0c0; font-family:lucida"><i>The Missing values in the <b>Brand</b> column are replaced by <b>TMA</b> using the .fillna() function</i></li>

<span style="color:#4169e1; font-family: copperplate; font-size: 26px; font-weight: bold">b. Handling Category Column:</span>

```python
recalldf['Category'] = recalldf['Category'].fillna("Accessories")
recalldf.head(6)
```

<li style="color:#c0c0c0; font-family:lucida"><i>The Missing values in the <b>Category</b> column are replaced by <b>Accessories</b> using the .fillna() function.</i></li>

<span style="color:#4169e1; font-family: copperplate; font-size: 26px; font-weight: bold">c. Handling Ratings Column:</span>

```python
#Calculating the median of the Rating column
rating_median = recalldf['Rating'].median()
print("Median of Rating column:", rating_median)
```

```python
# Filling the missing values with the median of the Rating Column
recalldf['Rating'] = recalldf['Rating'].fillna(rating_median)
recalldf.head(7)
```

<li style="color:#c0c0c0; font-family:lucida"><i>The Median of the 'Rating' column is calculated using the <b>.median()</b> function. The Median value calculated is <b>3.0.</b></i>i></li>
<li style="color:#c0c0c0; font-family:lucida"><i>The Missing values in the <b>Rating</b> column are replaced by <b>Median (3.0)</b> using the .fillna() function</i></li>

<span style="color:#4169e1; font-family: copperplate; font-size: 26px; font-weight: bold">d. Handling Return Rate Column:</span>

```python
rating_median = recalldf['Return_Rate'].median()
print("Median of Rating column:",rating_median)
```

```python
recalldf['Return_Rate'] = recalldf['Return_Rate'].fillna(rating_median)
recalldf.head(38)
```

<li style="color:#c0c0c0; font-family:lucida"><i>The Median of the 'Return_Rate' column is calculated using the <b>.median()</b> function. The Median value calculated is <b>0.15.</b></i>i></li>
<li style="color:#c0c0c0; font-family:lucida"><i>The Missing values in the <b>Return_Rate</b> column are replaced by <b>Median (0.15)</b> using the .fillna() function</i></li>

<span style="font-family: impact; font-size: 30px; color:#005b5b;"><u>VISUALIZATION</span>

<span style="color:#4169e1; font-family: copperplate; font-size: 25px; font-weight: bold">↪ <u><i>Pie Chart:</span>

<span style="color: coral; font-family: copperplate; font-size: 19px; font-weight: bold"><i>i) Recall Rate</span>

```python
recallFlag=recalldf['Recall_Flag'].value_counts().sort_index()
print(recallFlag)
```

<li style="color:#c0c0c0; font-family:lucida"><b>.value_counts().sort_index()</b><i> counts how often each unique value appears in a column and then sorts the results in order based on the values (not the counts).</i></li>

```python
plt.figure(figsize=(12,5))
recallFlag.plot(kind='pie',autopct='%1.1f%%')
plt.title('Recall Rate')
plt.show()
```

<li style="color:#c0c0c0; font-family:lucida"><b>plt.figure(figsize=(12,5)): </b><i>Sets the size of the figure to 12 inches wide and 5 inches tall.</i></li>

<li style="color:#c0c0c0; font-family:lucida"><b>.plot(kind='pie', autopct='%1.1f%%'): </b><i> Creates a pie chart from the recallFlag data and shows percentage values on each slice.</i></li>

<li style="color:#c0c0c0; font-family:lucida"><b>plt.title(''): </b><i> Adds a title to the chart.</i></li>

<li style="color:#c0c0c0; font-family:lucida"><b>plt.show():</b><i> Displays the chart.</li></i><br>



<li style="color:#c0c0c0; font-family:lucida"><i>The above Pie Chart displays that <b>94.5%</b> times the product is not recalled while <b>5.5%</b> times the product is recalled</i></li>

<span style="color: coral; font-family: copperplate; font-size: 19px; font-weight: bold"><i>ii) Product Availability</span>

```python
available=recalldf['Availability'].value_counts().sort_index()
print(available)
```

```python
plt.figure(figsize=(12,5))
available.plot(kind='pie', autopct='%1.1f%%', cmap='coolwarm')
plt.title('Rating')
plt.show()
```

<li style="color:#c0c0c0; font-family:lucida"><i>The above Pie Chart displays that <b>25.6%</b> of the products are <b>In Stock</b>, <b>25.2%</b> of the products are in <b>In Limited Stock</b>, <b>25.2%</b> of the products will be <b>Shipped In 3-5 Days</b> and <b>24.0%</b> of the products are unavailble.</i>

<span style="color:#4169e1; font-family: copperplate; font-size: 26px; font-weight: bold">↪ <u><i>Correlation Matrix:</span>

```python
col=recalldf.select_dtypes(include=['int64','float64'])
#selecting num col
cor_mat=col.corr() 
```

```python
#Visualizing Heat Map
sns.heatmap(cor_mat,annot=True,fmt='.2f',cmap='coolwarm',linewidths=0.5) 
plt.title("Correlation Heatmap")
```

<p style="color:#c0c0c0; font-family:lucida"><i>This is a Correlation Heatmap, showing how strongly different numeric features (columns) are related to each other in your dataset.<BR>
Values range from -1 to 1.</i></p>

<li style="color:#c0c0c0; font-family:lucida"><i>A value near 1 = strong positive correlation (they increase together).</i></li>
<li style="color:#c0c0c0; font-family:lucida"><i>A value near -1 = strong negative correlation (one increases, the other decreases).</i></li>
<li style="color:#c0c0c0; font-family:lucida"><i>A value near 0 = no or very weak correlation.</i></li>
